# Baixando e configurando o Spark

Passos:

1º - Baixe o Spark pelo site: https://spark.apache.org/downloads.html


2º - Salve o arquivo descompactado em "Documentos"


3º - Baixe o Winutils e o hadoop.dll pelo site: https://github.com/kontext-tech/winutils

** ATENÇÃO: AO BAIXAR O WINUTILS E HADOOP.DLL SE ATENTAR A VERSÃO. ELAS PRECISAM SER COMPATÍVEIS **


4º - Baixe o java JDK pelo site: https://www.oracle.com/br/java/technologies/javase/javase8-archive-downloads.html


5º - Configure as varíaveis de ambiente(ENV)

    Variáveis de usuário:
    SPARK_HOME > C:\User\Documents\VersãoSpark
    HADOOP_HOME > %SPARK_HOME%\hadoop
    JAVA_HOME > C:\Program Files\Java\VersãoJDK
    PYSPARK_DRIVER_OPTS > notebook
    PYSPARK_DRIVER_PYTHON > jupyter
    PYSPARK_PYTHON > python

    Vavariável do sistema:
    Adicionar ao Path > 
        %HADOOP_HOME%\bin
        %SPARK_HOME_%\bin


6º - Criar Ambiente no Python  

    Python:Create Enviroment > Venv Creates > Python
    Aguardar o processo de criação


7º - Instalar o PySpark

    %pip install pyspark=="SuaVersão"


8º - Selecionar o kernel criado


9º - Baixar JARs necessários atraves do site https://mvnrepository.com/

    hadoop-aws
    aws-java-sdk-bundle
    delta-storage
    delta-spark

** ATENÇÃO: AO BAIXAR O WINUTILS E HADOOP.DLL SE ATENTAR A VERSÃO. ELAS PRECISAM SER COMPATÍVEIS **


10º - Executar Script "Delta_Spark.py"
